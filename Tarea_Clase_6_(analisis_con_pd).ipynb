```python
#Cargamos df "Mcdonalds" que contiene el analisis de los valores nutricionales de una variedad de comidas y snacks.
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/mcdonalds.xlsx')
df.head()
```

Sin tener el detalle de cual es la unidad de medida con la cual se miden los macronutrientes y otros indicadores nutricionales se asume que son en gramos.

```python
#Empezamos con lo básico, la dimensión/tamaño de nuestra base de datos
df.shape
#Para este df, tenemos la información nutricional de 121 items
```

```python
#Buscamos un poco más de detalle y vemos que el df cuenta con 8 categorias de alimentos, donde predomina aquellos categorizados como Desayunos, Pollos y pescados y Carne de cerdo.
df.groupby(['categoria'])['articulo_id'].count()
```

```python
df_macronutrientes = df.groupby(['categoria'])[['tamano_porcion','calorias','grasa_total','proteina','carbohidratos']].mean()
df_macronutrientes.sort_values(by=['calorias','tamano_porcion'], ascending=False)
#Agrupamos la media de los macronutrientes más importantes de cada una de las categorias para tener una idea del aporte nutricional que tiene cada uno
```

Tomamos la tabla y la ordenamos las categorias según la cantidad de calorias que tiene cada categoria de forma descendiente. Con esto podemos armar un ratio entre porción y caloria para ver que categoria de alimentos son de porciones grandes pero son bajos en calorias.

---



```python
df_macronutrientes['ratio_porcion_calorias'] = df_macronutrientes['tamano_porcion'] / df_macronutrientes['calorias']
df_macronutrientes.sort_values(by=['ratio_porcion_calorias'], ascending = False)
```

Con este pequeño analisis podemos ver que las bebidas parecieran ser la categoria más "eficiente" a nivel de porción y calorias. Una porción promeido de bebida contiene menos calorias que las otras categorias, luego aparecen las ensaladas lo cual tiene sentido si lo comparamos con un desayuno o algún plato categorizado como "carne de cerdo" o "pollo y pescado".
